# Conectando e fazendo transferência para o Postgres

### Aqui foi feito a conexão com o SGBD do Postgres e feito a criação das tabelas e feito a transferência para o SGBD para análise e criação das tabelas para insights e criação dos gráficos que serão utilizados no PowerBI

```python
# Libs
import pandas as pd
import os

# Carregar modulos
from e_commerce.data_extraction.extractor import get_data
from e_commerce.conections.sql import test_connection
from sqlalchemy import create_engine, text
from dotenv import load_dotenv

# Carrega load_dotenv
load_dotenv()
```

```python
# Credenciais do PostgreSQL
usuario_pg = os.getenv("POSTGRES_USER")
senha_pg = os.getenv("POSTGRES_PASSWORD")
host_pg = os.getenv("POSTGRES_HOST")
porta_pg = os.getenv("POSTGRES_PORT")
banco_pg = os.getenv("POSTGRES_DB")
schema = os.getenv("POSTGRES_SCHEMA")
```

```python
df_sql = get_data(source_type='csv', type_name='processed', filename_or_path='e_commerce_prepared.csv')

```

```python
df_sql
```

```python
# Criar engine com banco 
engine = create_engine(f"postgresql+psycopg2://{usuario_pg}:{senha_pg}@{host_pg}:{porta_pg}/{banco_pg}",
                        connect_args={'options': '-c search_path="e-commerce"'})

# Testar a conexão
test_connection(engine, schema)
```

```python
# Confirma em qual schema do DB você está
with engine.connect() as conn:
    resultado = conn.execute(text("SHOW search_path;")).fetchall()
    print(resultado)
```

```python
df_sql.to_sql('ecommerce_america', engine, if_exists='replace', index=False)
```
